```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
def gausienne(x,t):
    return np.exp(-(np.pow(x,2)) / t)
```

```python
X_lims = np.array([-5, 5])
t = 3
# Polynôme de Lagrange (ordre 2 puisque 3 points)


LEVEL = 10
N_PTS = 2**LEVEL
TARGET = .83
dx = (X_lims[1] - X_lims[0])/N_PTS
target_index = int(TARGET * N_PTS)
# Évaluation sur une grille fine
x_vals = np.linspace(min(X_lims), max(X_lims), N_PTS)
y_reals = gausienne(x_vals,t)

X_SET = x_vals[target_index-1:target_index+2]
Y_SET = y_reals[target_index-1:target_index+2]
fig,axs = plt.subplots(1,2,figsize = (15,7),sharey=True)
axs[0].scatter(x_vals[target_index],y_reals[target_index])
axs[0].plot(x_vals, y_reals, "-")
poly = lagrange(X_SET,Y_SET)
x_thin = np.linspace(min(X_SET) - 1*dx, max(X_SET) +1*dx, N_PTS)
y_pred_thin = poly(x_thin)
axs[1].scatter(X_SET ,  Y_SET)
plt.plot(x_thin , y_pred_thin ,color = "navy",linewidth = 10,alpha = .4)
plt.plot(x_thin , gausienne(x_thin,t) ,color = "red")
plt.show()
```
